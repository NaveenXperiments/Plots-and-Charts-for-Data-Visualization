```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = [iris.target_names[target] for target in iris.target]

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=iris_df,
    x=iris.feature_names[2],  # Petal length
    y=iris.feature_names[3],  # Petal width
    hue='species',
    palette='viridis'
)
plt.title('Scatter Plot of Petal Length vs. Petal Width', fontsize=14)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.legend(title='Species', fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = [iris.target_names[target] for target in iris.target]

# Calculate mean values for each species
mean_values = iris_df.groupby('species').mean().reset_index()

# Line plot
plt.figure(figsize=(8, 6))
for feature in iris.feature_names:
    sns.lineplot(
        data=mean_values,
        x='species',
        y=feature,
        marker='o',
        label=feature
    )

plt.title('Mean Feature Values Across Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Mean Value', fontsize=12)
plt.legend(title='Features', fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = [iris.target_names[target] for target in iris.target]

# Calculate mean values for each species
mean_values = iris_df.groupby('species').mean().T  # Transpose for area plot

# Area plot
plt.figure(figsize=(10, 6))
for column in mean_values.columns:
    plt.fill_between(mean_values.index, mean_values[column], label=column, alpha=0.6)

plt.title('Area Plot of Mean Feature Values by Species', fontsize=14)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Mean Values', fontsize=12)
plt.legend(title='Species', fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Stacked plot
plt.figure(figsize=(10, 6))
plt.stackplot(
    mean_values.index,
    mean_values.iloc[:, 0],  # Setosa
    mean_values.iloc[:, 1],  # Versicolor
    mean_values.iloc[:, 2],  # Virginica
    labels=mean_values.columns,
    alpha=0.7
)

plt.title('Stacked Area Plot of Mean Feature Values by Species', fontsize=14)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Mean Values', fontsize=12)
plt.legend(title='Species', fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Count the number of samples per species
species_counts = df['species'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    species_counts,
    labels=species_counts.index,
    autopct='%1.1f%%',  # Display percentages
    startangle=140,     # Rotate start angle
    colors=['#ff9999', '#66b3ff', '#99ff99']  # Custom colors
)
plt.title('Distribution of Iris Species', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = [iris.target_names[target] for target in iris.target]

# Calculate mean values for each species
mean_values = iris_df.groupby('species').mean()

# Create the table chart
fig, ax = plt.subplots(figsize=(8, 4))
ax.axis('off')  # Turn off the axis
table = ax.table(
    cellText=mean_values.round(2).values,  # Rounded values for better display
    colLabels=mean_values.columns,        # Column labels (features)
    rowLabels=mean_values.index,          # Row labels (species)
    cellLoc='center',                     # Center align text in cells
    loc='center'                          # Position the table at the center
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width(col=list(range(len(mean_values.columns))))

plt.title('Mean Feature Values for Each Iris Species', fontsize=14, pad=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the polar chart
categories = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
values = [5.84, 3.05, 3.76, 1.20]  # Example mean values for the Iris dataset

# Create the polar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
values += values[:1]  # Repeat the first value to close the circular graph
angles += angles[:1]  # Repeat the first angle

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=1)
ax.plot(angles, values, color='blue', linewidth=2)
ax.set_yticks([1, 2, 3, 4, 5])  # Optional: adjust radial ticks
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title('Polar Chart of Iris Features', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris_df = sns.load_dataset('iris')

# Calculate the average petal length by species
average_petal_length = iris_df.groupby('species')['petal_length'].mean()

# Plotting the lollipop chart
plt.figure(figsize=(8, 5))
plt.plot(average_petal_length, 'bo-', markersize=10)  # 'bo-' means blue circles with a solid line
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.title('Lollipop Chart of Average Petal Length by Species')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the lollipop chart
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
values = [5.84, 3.05, 3.76, 1.20]  # Example mean values for the Iris dataset

# Create the lollipop chart
fig, ax = plt.subplots(figsize=(8, 5))
ax.stem(features, values, basefmt=" ", use_line_collection=True)
ax.scatter(features, values, color='red', s=100, zorder=2)  # Add markers
ax.set_ylabel('Mean Values', fontsize=12)
ax.set_title('Lollipop Chart of Iris Features', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
